```python
import pandas as pd
import sqlite3
```

```python
PuertoRican_beaches= pd.read_csv('Monitoreo de Playas 2018')
PuertoRican_beaches.head()
```

```python
PuertoRican_beaches=PuertoRican_beaches.rename(columns={'Beach ID':'beach_id', 'Beach Name':'beach_name','Sampling Route':'sampling_route', 'Sampling Type':
                        'sampling_type',
       'Enterococcus\r\r\n (colonies/100 mL)': 'enterococcus_colonies', 'Temperature\r\r\n(°C)': 'temperature'} )
```

```python
#tool that enables you to connect to the database
conn=sqlite3.connect('beaches.db')
```

```python
PuertoRican_beaches.to_sql('PRbeaches',conn)
```

```python
pd.read_sql("SELECT * FROM PRbeaches",conn)
```
